#**Audio Visualizations**

```python
import librosa.display
import librosa.display as lplt

plt.figure(figsize=(12,4))
librosa.display.waveshow(data, color = "Blue")
plt.figure(figsize=(12,4))
librosa.display.waveshow(data2, color = "Blue")
plt.show()
```

```python
#Spectrogram
stft = librosa.stft(data)
stft_db = librosa.amplitude_to_db(abs(stft))
plt.figure(figsize=(12,4))
librosa.display.specshow(stft, sr=samplingrate, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
#stft the amplitude to db
stft = librosa.stft(data)
stft_db = librosa.amplitude_to_db(abs(stft))
plt.figure(figsize=(12,4))
librosa.display.specshow(stft_db, sr = samplingrate, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
#computing the rolloff frequency for each frame in a signal
spectral_rolloff = librosa.feature.spectral_rolloff(data+0.01, sr=samplingrate)[0]
plt.figure(figsize=(12,4))
librosa.display.waveshow(data, sr=samplingrate, alpha=0.4, color='Purple')
```

```python
#Chroma feature or chromagram closely relates to the twelve different pitch classes.
#Chroma features capture harmonic & melodic characteristics of music while being robust to changes.
chroma = librosa.feature.chroma_stft(data, sr=samplingrate)
plt.figure(figsize=(16,6))
lplt.specshow(chroma, sr=samplingrate, x_axis='time', y_axis='chroma', cmap='BuPu')
plt.colorbar()
plt.title("Chroma Features")
plt.show()
```

```python
#The rate at which zero-crossings occur is a simple measure of the frequency content of a signal.
#Zero crossing is said to occur if successive samples have different algebric signs.
start = 1000
end = 1200
plt.figure(figsize=(14,5))
plt.plot(data[start:end], color='Purple')
plt.grid()
```

```python
zero_cross_rate = librosa.zero_crossings(data[start:end], pad=False)
print("The number of zero crossings is:",sum(zero_cross_rate))
```
