#**Understanding Audio files**

```python
import librosa
import matplotlib.pyplot as plt
```

```python
audio= '/content/drive/MyDrive/music_dataset/music_dataset/bati/Bat000.wav'
audio2 = '/content/drive/MyDrive/music_dataset/music_dataset/ambasel/amb040.wav'
audio3= '/content/drive/MyDrive/music_dataset/music_dataset/tizita/tzt000.wav'
audio4 = '/content/drive/MyDrive/music_dataset/music_dataset/anchihoye/Anch000.wav'
```

```python
# using librosa to load the audio file
data , samplingrate = librosa.load(audio)
data2, samplingrate2 = librosa.load(audio2)
data3 , samplingrate3 = librosa.load(audio3)
data4, samplingrate4= librosa.load(audio4)
```

#**Audio Visualizations**

```python
import librosa.display
import librosa.display as lplt

#plotting the changes in the amplitude of a sound over time

plt.figure(figsize=(12,4))
librosa.display.waveshow(data, color = "Blue", label = "BATI")
plt.figure(figsize=(12,4))
librosa.display.waveshow(data2, color = "Blue", label = "AMBASEL")
plt.figure(figsize=(12,4))
librosa.display.waveshow(data3, color = "Blue", label = "TIZITA")
plt.figure(figsize=(12,4))
librosa.display.waveshow(data4, color = "Blue", label = "ANCHIHOYE")
plt.show()
```

```python
#Spectrogram
#Amplitude -  on the decibel (dB) scale, measures sound's pressure or forcefulness

stft = librosa.stft(data)
stft_db = librosa.amplitude_to_db(abs(stft))
plt.figure(figsize=(12,4))
librosa.display.specshow(stft, sr=samplingrate, x_axis='time', y_axis='hz')
plt.colorbar()
```

**Rolloff Frequency**


```python
#rolloff frequency
# the frequency under which some part of the total energy of the spectrum is contained.

spectral_rolloff = librosa.feature.spectral_rolloff(data+0.01, sr=samplingrate)[0]
plt.figure(figsize=(12,4))
librosa.display.waveshow(data, sr=samplingrate, alpha=0.4, color='Purple')

spectral_rolloff = librosa.feature.spectral_rolloff(data3+0.01, sr=samplingrate)[0]
plt.figure(figsize=(12,4))
librosa.display.waveshow(data3, sr=samplingrate, alpha=0.4, color='Purple')
```

**Chroma**

```python
#Chroma features capture harmonic & melodic characteristics of music while being robust to changes.

chroma = librosa.feature.chroma_stft(data, sr=samplingrate)
plt.figure(figsize=(25,6))
lplt.specshow(chroma, sr=samplingrate, x_axis='time', y_axis='chroma', cmap='BuPu')
plt.colorbar()
plt.title("Chroma Features")
plt.show()
```

```python
#The rate at which zero-crossings occur is a simple measure of the frequency content of a signal.
#Zero crossing is said to occur if successive samples have different algebric signs.

plt.figure(figsize=(14,5))
plt.plot(data2[1000:1200], color='Blue')
plt.plot(data4[1000:1200], color='Green')
plt.grid()
```

```python

for i in [data2,data4]:
  zero_cross_rate = librosa.zero_crossings(i[1000:1200], pad=False)
  print("The number of zero crossings is:",sum(zero_cross_rate))
```
